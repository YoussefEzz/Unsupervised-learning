```python
import pandas as pd


#read and parse the training set .csv features file 
df = pd.read_csv('A3-data.txt', delimiter = ',', usecols=['x', 'y', 'z', 't'])
class_label = pd.read_csv('A3-data.txt', delimiter = ',', usecols=['class'])

df.head()
```

```python
#read four input(features) columns
x = df[df.columns[0]]
y = df[df.columns[1]]
z = df[df.columns[2]]
t = df[df.columns[3]]

data = [x, y, z, t]
print(data)
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

#create tsne model for 2 tsne dimensions
# perplexity : allowing ditances to be grouped together
tsne = TSNE(n_components=2, perplexity=20, n_iter=1000, random_state=0 )

tsne_components = tsne.fit_transform(df)
#print(tsne_components)

tsne_df = pd.DataFrame(tsne_components,columns = ['tsne1', 'tsne2'])
tsne_df.head()

tsne_df = pd.concat([tsne_df, class_label], axis=1)
tsne_df.head()

print(tsne_df)
#scatter plot of the first two tsne componenets with labels
plt.scatter(tsne_df['tsne1'], tsne_df['tsne2'], c = tsne_df['class'])
plt.xlabel("tsne1")
plt.ylabel("tsne2")
plt.title("t-SNE projection in two dimensions with perplexity = 20")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

#create tsne model for 2 tsne dimensions
# perplexity : allowing ditances to be grouped together
tsne = TSNE(n_components=2, perplexity=50, n_iter=1000, random_state=0 )

tsne_components = tsne.fit_transform(df)
#print(tsne_components)

tsne_df = pd.DataFrame(tsne_components,columns = ['tsne1', 'tsne2'])
tsne_df.head()

tsne_df = pd.concat([tsne_df, class_label], axis=1)
tsne_df.head()

print(tsne_df)
#scatter plot of the first two tsne componenets with labels
plt.scatter(tsne_df['tsne1'], tsne_df['tsne2'], c = tsne_df['class'])
plt.xlabel("tsne1")
plt.ylabel("tsne2")
plt.title("t-SNE projection in two dimensions with perplexity = 50")
```
