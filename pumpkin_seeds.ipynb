```python
import pandas as pd
# https://www.kaggle.com/datasets/muratkokludataset/pumpkin-seeds-dataset
df = pd.read_excel('Datasets/Pumpkin_Seeds_Dataset.xlsx', sheet_name='Pumpkin_Seeds_Dataset')
df.head()
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Print the target variable values, and its counts. 
target_value_count = df['Class'].value_counts() 
print(target_value_count)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Class'] = label_encoder.fit_transform(df['Class'])
df.head()
```

```python
import matplotlib.pyplot as plt

# Extract the columns for the boxplot
# Get the list of column names
column_names = df.columns.tolist()
print(len(column_names))


# Get the list of lists of column data
column_data = df.values
print((column_data))

fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(column_data,
                notch=True,  # notch shape
                vert=True,  # vertical box alignment
                patch_artist=True,  # fill with color
                labels=column_names)
 
# show plot
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((0,1))
min_max_scaling_columns = ["Area",	"Perimeter",	"Major_Axis_Length",	"Minor_Axis_Length",	"Convex_Area",	"Equiv_Diameter"]



df[min_max_scaling_columns] = scaler.fit_transform(df[min_max_scaling_columns])
df.head()
```

```python
# Get the list of column names
column_names = df.columns.tolist()
print(len(column_names))


# Get the list of lists of column data
column_data = df.values
print((column_data))

fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(column_data,
                notch=True,  # notch shape
                vert=True,  # vertical box alignment
                patch_artist=True,  # fill with color
                labels=column_names)
 
# show plot
plt.show()
```

```python
# Write normalized DataFrame to a table-like format (CSV file)
df.to_csv('Datasets\Pumpkin_Seeds_Normalized.csv', index=False, sep='\t')
```
